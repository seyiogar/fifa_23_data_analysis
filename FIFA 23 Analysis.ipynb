# Importing Libraries and reading data 

```python
#Importing necessary libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
pd.set_option('display.max_columns', 89)
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv(r'C:\Users\hp\Documents\Fifa 23 Players Data.csv\Fifa 23 Players Data.csv')
df.head()
```

# Data Cleaning and Preparation

```python
df.info()
```

```python
df.rename(columns = {' GoalkeeperKicking': 'Goalkeeper Kicking', 'LongPassing': 'Long Passing','BallControl': 'Ball Control', 'TotalStats': 'Total Stats', 'BaseStats':'Base Stats'  }, inplace=True)
df.columns
```

```python
df.columns = df.columns.str.rstrip(')')
df.columns = df.columns.str.replace('(',' ')
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df.describe()
```

```python
#Dropping Uneccesary columns 
fifa = df.drop(columns = ['Image_Link','National_Team_Image_Link'], axis=1)
```

```python
fifa.head(10)
```

```python
# Checking for records that contain missing values
fifa[fifa.isnull().any(axis = 1)]
```


# Exploratory Analysis and Visualization


```python
fifa.head()
```

```python
# Overall rating distribution 
plt.figure(figsize=(15,8))
sns.distplot(fifa['Overall'], bins=15, color='r') 
plt.title('Overall Rating Distribution in FIFA 23', fontsize = 16)
plt.show()
```

```python
# compute correlation matrix of the relevant columns 
relevant_col = fifa[['Overall', 'Potential', 'Value_in_Euro', 'Age','Height_in_cm', 'Weight_in_kg', 'Wage_in_Euro','Release_Clause']].corr()
corr = relevant_col.corr()
corr
```

```python
# Display correlation heatmap
plt.figure(figsize=(15,10))
mask= np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, mask=mask, cmap= 'vlag', center=0)
```

## Q) Who are the oldest and youngest players in FIFA 23?

```python
# The oldest player
oldest_plyr =fifa[['Full_Name', 'Age']].loc[fifa['Age'] == fifa['Age'].max()]

oldest_plyr
```

```python
#The youngest players
youngest_plyr = fifa[['Full_Name', 'Age']].loc[fifa['Age'] == fifa['Age'].min()]

youngest_plyr
```

## Q) What is the median age of a player in FIFA 23? 

```python
# The median age of a player 
med_age = fifa['Age'].median()
med_age
```

## Q) What is the median overall rating of a player in FIFA 23? 

```python
# The median Overall rating of a player
med_overall = fifa['Overall'].median()
med_overall
```

## Q) Who are the  top 10 overall players in FIFA 23? 

```python
# The Top 10 ranked by Overall rating 
top_ten_overall = fifa[['Full_Name','Overall','Age']].nlargest(10, 'Overall')

top_ten_overall
```

## Q) Who are the top 20 most valuable players in FIFA 23?

```python
# Top 20 most valuable players
top_ten_val = fifa[['Known_As','Full_Name','Value_in_Euro','Age','Overall']].nlargest(20, 'Value_in_Euro')
top_ten_val
```

```python
plt.figure(figsize=(15,7))
sns.barplot(top_ten_val['Known_As'], top_ten_val['Value_in_Euro'], color = 'm')
plt.tick_params(axis = 'x', rotation=60)
plt.title('Most Valuable Players in FIFA 23', fontsize= 15)
plt.ylabel('Value in Euro', fontsize= 15)
plt.show()
```

## Q) Who is the highest paid player in FIFA 23?

```python
fifa[['Full_Name','Wage_in_Euro']].loc[fifa['Wage_in_Euro'] == fifa['Wage_in_Euro'].max()]
```

## Q) Which player has the highest potential in FIFA 23?

```python
#The best potential player 
best_pot_plyr = fifa[['Full_Name', 'Potential','Age']].loc[fifa['Potential'] == fifa['Potential'].max()]

best_pot_plyr
```

```python
# Age vs Overall rating
plt.figure(figsize=(15,8))
sns.scatterplot(x = 'Age', y = 'Overall', data = fifa)
plt.title('Age vs Overall Rating', fontsize = 15)
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Overall Rating', fontsize = 15)
```

## Q) What's the count of players on the basis of age?

```python
#Count of Players on Basis of Age
fifa['Age'].value_counts()
```

```python
plt.figure(figsize =(15,7))
sns.countplot(x = 'Age', data = fifa)
plt.title('Count of Players on the Basis of Age (FIFA 23)', fontsize = 16)
plt.xlabel('Age', fontsize = 16)
plt.ylabel('Count', fontsize = 16)
```

## Q) What are the top 20 countries with the most players in FIFA 23?

```python
# Top 20 Countries with the most amount of players 
top_countries = fifa['Nationality'].value_counts().head(20)
top_countries
```

```python
plt.figure(figsize =(15,7))
sns.barplot(x = top_countries.index, y = top_countries, color='k')
plt.tick_params(axis = 'x', rotation= 60)
plt.title('Top Countries', fontsize = 15)
plt.ylabel('Number of Players', fontsize = 15)
```
